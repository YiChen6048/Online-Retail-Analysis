```python
df.loc[:, 'Description'] = df['Description'].str.lower().str.strip()
```

📌why don't let me use df['Description'] = df['Description'].str.lower().str.strip()?

Ans:
你可能在对原始 DataFrame 的一个“切片（slice）”做修改，而这个“切片”可能是原始数据的副本（copy），不是视图（view），所以你对它的修改可能不会影响到原始数据

🔍 警告出现的根本原因：
你之前可能有类似这样的操作：

df = original_df[some_condition]  # 比如 df = df[df['Quantity'] > 0]

然后你再对 df['Description'] 做修改：

df['Description'] = df['Description'].str.lower().str.strip()

这时候 pandas 不确定 df 是原始表的一部分（视图），还是复制出来的新表（副本），所以发出 SettingWithCopyWarning。

✅ 正确做法：用 .loc 明确告诉 pandas 你要修改哪一部分, pandas 会尽可能在“视图”上修改（也就是原始数据的一个窗口）；
df.loc[:, 'Description'] = df['Description'].str.lower().str.strip()

✅ 更稳妥的写法（先 copy 再操作）：
如果你确定要操作的是一个副本，那可以显式 .copy() 来避免歧义：

df = df[df['Quantity'] > 0].copy()

df['Description'] = df['Description'].str.lower().str.strip()

这样就不会出现警告了。

```python
df['Description'].str.contains('gift', na=False)
```

📌why do I need to add na=False?

Ans:
如果某一行 Description 是缺失值（NaN），那结果将是 NaN，不是 True 或 False，这在后续做筛选时可能导致报错或不确定的行为。

⚠️NaN 出现的话，你会得到一个错误：

ValueError: cannot mask with non-boolean array containing NA / NaN values

✅ 加上 na=False 的作用，就会把 NaN 自动当作 False 处理，这样你就可以安全地筛选了：

df[df['Description'].str.contains('gift', na=False)]

就不会报错。

```python
# 使用向量化方法判断高价值订单
df['HighValueCustomer'] = df['TotalPrice'] > 1000
```

```python
high_value_df = df[df['HighValueCustomer']]
```

```python
print(high_value_df)
```

```python
print("高价值交易数量：", df['HighValueCustomer'].sum())
```

用 Python for 循环实现：

#假设你已有 df，且包含 TotalPrice 列

high_value_list = []  # 用来存放每条记录是否为高价值客户

for price in df['TotalPrice']:

    if price > 1000:
    
        high_value_list.append(True)
        
    else:
    
        high_value_list.append(False)

#把结果赋值回新的列

df['HighValueCustomer'] = high_value_list

#查看前几条结果

print(df[['TotalPrice', 'HighValueCustomer']].head())

如果你的数据量大，比如上万、几十万条，这个循环会运行得很慢。

向量化写法一行代码搞定，速度快而且更优雅
